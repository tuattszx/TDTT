```python
def bs(arr : list, target : int) ->int:
    l, r = 0, len(arr) - 1
    while l < r:
        m = (l + r) >> 1
        if arr[m] < target:
            l = m + 1
        else:
            r = m
    return r if arr[r] == target else -1
arr = list(map(int, input().split()))
target = int(input())
print(bs(arr, target))
```

```python
def count(arr : list, x : int) ->int:
    cnt = 0
    for t in arr:
        cnt += t == x
    return cnt
arr = list(map(int, input().split()))
x = int(input())
print(count(arr, x))
```
