```python
def bs(arr : list, target : int) ->int:
    l, r = 0, len(arr) - 1
    while l < r:
        m = (l + r) >> 1
        if arr[m] < target:
            l = m + 1
        else:
            r = m
    return r if arr[r] == target else -1
arr = list(map(int, input().split()))
target = int(input())
print(bs(arr, target))
```

```python
def Bubble_Sort(arr : list) -> list:
    for i in range(len(arr)):
        for j in range(len(arr) - i - 1):
            if arr[j] > arr[j + 1]:
                arr[j], arr[j + 1] = arr[j + 1], arr[j]
def Selection_Sort(arr : list) -> list:
    for i in range(len(arr)):
        min_idx = i
        for j in range(i + 1, len(arr)):
            if arr[j] < arr[min_idx]:
                min_idx = j
        arr[i], arr[min_idx] = arr[min_idx], arr[i]
arr = list(map(int, input().split()))
Bubble_Sort(arr)
print(arr)
Selection_Sort(arr)
print(arr)
```

```python
def f(arr : list) -> list:
    dt = {}
    #x, idx, cnt
    for i in range(len(arr)):
        try:
            dt[arr[i]][1] += 1
        except:
            dt[arr[i]] = [i, 1]
    res = [0, 0, 0]
    for key, val in dt.items():
        if (val[1] > res[2]):
            res = [key, val[0], val[1]]
    return (res[0], res[2])
arr = list(map(int, input().split()))
print(f(arr))
```

```python
def f(arr : list, x : int)->int:
    cnt = 0
    for i in range(len(arr)):
        for j in range(i + 1, len(arr)):
            if (arr[i] + arr[j]) - x == 0:
                cnt += 1
    return cnt
arr = list(map(int, input().split()))
x = int(input())
print(f(arr, x))
```

```python
from bisect import bisect_left, bisect_right
INF = int(1e18)
n, m = map(int, input().split())
L, R = [0] * n, [0] * n
for i in range(n):
    L[i], R[i] = map(int, input().split())
Q = [int(input()) for i in range(m)]
res = [INF] * m
for i in range(n):
    x = R[i] - L[i] + 1
    l = bisect_left(Q, L[i])
    r = bisect_right(Q, R[i])
    for j in range(l, r):
        res[j] = min(res[j], x)
for i in range(m):
    res[i] = res[i] if res[i] != INF else -1
print(res)
```

```python
class Node:
    def __init__(self, key : int):
        self.key = key
        self.left = None
        self.right = None
class BST:
    def __init__(self):
        self.root = None
    def insert(self, key : int):
        if (self.root == None):
            self.root = Node(key)
        else:
            self.dfs(self.root, key)
    def dfs(self, node : Node, key : int):
        if key < node.key:
            if node.left == None:
                node.left = Node(key)
            else:
                self.dfs(node.left, key)
        else:
            if node.right == None:
                node.right = Node(key)
            else:
                self.dfs(node.right, key)
    def count(self, node : Node, x : int):
        if node is None:
            return 0       
        if node.key < x:
            return 1 + self.count(node.left, x) + self.count(node.right, x)
        else:
            return self.count(node.left, x)

a = list(map(int, input().split()))
n = len(a)
res = [0] * n
tree = BST()
for i in range(n - 1, -1, -1):
    res[i] = tree.count(tree.root, a[i])
    tree.insert(a[i])
print(res)
```
