```python
def Max_Of_Two(a, b):
    return a if a > b else b
a, b = map(int, input().split())
print(Max_Of_Two(a, b))
```

```python
def prime(n):
    for i in range(2, int(n ** 0.5) + 1):
        if n % i == 0:
            return False
    return n > 1
n = int(input())
print(f"{n} is prime" if prime(n) else f"{n} is not prime")
```

```python
def so_hh(n)->bool:
    s = 0
    for i in range(1, n // 2 + 1):
        if n % i == 0:
            s += i
    return s == n
n = int(input())
print(so_hh(n))
```

```python
def find(arr, k):
    for i in range(len(arr)):
        if arr[i] == k:
            return i + 1
    return -1
arr = list(map(int, input().split()))
k = int(input())
print(find(arr, k))
```

```python
def f(n):
    if n <= 1:
        return 1
    return n * f(n - 1)
n = int(input())
print(f(n))
```

```python
def haming(a, b):
    x = a ^ b
    res = 0
    while x > 0:
        res += x & 1
        x >>= 1
    return res
a, b = map(int, input().split())
print(haming(a, b))
```

```python
def sum_digit(n):
    s = 0
    while n > 0:
        s += n % 10
        n //= 10
    return s
n = int(input())
print(sum_digit(n))
```

```python
def f(a, b)->bool:
    ds = dict()
    for i in range(len(a)):
        if (ds.get(a[i]) == None):
            ds[a[i]] = b[i]
        else:
            if (ds[a[i]] != b[i]):
                return False
    return True
a, b = input().split()
print(f(a, b) and f(b, a))
```
