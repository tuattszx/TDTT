```python
def g(s : str) ->  list:
    return s.split(':')
def f(arr : str) -> dict:
    res = dict()
    for s in arr:
        key, val = g(s)
        try:
            res[key].append(val)
        except:
            res[key] = [val]
    return res
arr = map(str, input().split())
print(f(arr))
```

```python
def g(s : str) ->  list:
    return s.split(':')
def h(s : str) -> dict:
    dt = dict()
    for t in s:
        key, val = g(t)
        val = int(val)
        try:
            dt[key] += val
        except:
            dt[key] = val
    return dt
def f(dt1 : dict, dt2 : dict) -> dict:
    for key, val in dt2.items():
        try:
            dt1[key] += val
        except:
            dt1[key] = val
    arr = list(dt1.keys())
    arr.sort()
    res = dict()
    for x in arr:
        res[x] = dt1[x]
    return res
s1 = map(str, input().split())
s2 = map(str, input().split())
print(f(h(s1), h(s2)))
```

```python
def f(arr : list, k : int) -> list:
    arr.sort()
    res = []
    for i in range(len(arr)):
        for j in range(i + 1, len(arr)):
            if (arr[i] + arr[j] == k):
                res.append((arr[i], arr[j]))
                break
    return res
arr = list(map(int, input().split()))
k = int(input())
print(f(arr, k))
```

```python
def f(arr : list) -> int:
    res = cnt = 0
    dt = dict()
    for i in arr:
        try:
            dt[i] += 1
        except:
            dt[i] = 1
        if (dt[i] == cnt):
            res = min(res, i)
        elif (dt[i] > cnt):
            cnt = dt[i]
            res = i
    return res
print(f(list(map(int, input().split()))))
```

```python
def f(a : list, b : list) -> list:
    res = []
    for x in a:
        if x in b and x not in res:
            res.append(x)
    return res
a = list(map(int, input().split()))
b = list(map(int, input().split()))
print(f(a, b))
```
